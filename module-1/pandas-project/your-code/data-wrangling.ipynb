```python
data_good_country=data[(data['Country'] == 'USA') | (data['Country'] == 'AUSTRALIA') | (data['Country'] == 'SOUTH AFRICA') ]
data_good_country.head()
```

```python
data_area=data_good_country.groupby('Area').count()
data_area
```

```python
areas=[i for i in data_area[data_area['Case Number'] >100].transpose()]
# data_area_usa=data_area_usa[(data_area_usa['Country'] == 'USA') | (data_area_usa['Country'] == 'AUSTRALIA') | (data['Country'] == 'SOUTH AFRICA') ]
print(areas)
```

```python
# data_good_country.drop(data_good_country[(data_good_country.Area != 'Florida') & (data_good_country.Area != 'California')].index)
# res=''
# for k,i in enumerate(areas):
#     if k != 0:
#         res+=' & '
#     res+='(data_good_country.Area != ' + '\''+ i +'\'' +')'
# print (res)
data_country_area=data_good_country.drop(data_good_country[(~data_good_country.Area.isin(areas))].index)
data_country_area
```

```python
data_country_area.groupby('Year').count().head()
```

```python
data_country_area=data_country_area.drop(data_country_area[(data_country_area.Year < 1900)].index)
data_country_area.groupby('Fatal (Y/N)').count()
```

```python
if list(data_country_area[data_country_area['Fatal (Y/N)'] == 'N '].index):
    a=list(data_country_area[data_country_area['Fatal (Y/N)'] == 'N '].index)[0]
    data_country_area.loc[a, 'Fatal (Y/N)']='N'

for i in list(data_country_area[data_country_area['Fatal (Y/N)'] == ' N'].index):
    data_country_area.loc[i, 'Fatal (Y/N)']='N'
data_country_area['Fatal (Y/N)'].value_counts()
# data_country_area[data_country_area['Fatal (Y/N)'] == 'N '].index
```
