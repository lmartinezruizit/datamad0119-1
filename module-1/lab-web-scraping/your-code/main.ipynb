####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")

table=soup.select('table')[3]
values = []
rows = table.select('tr')
for row in rows:
    cells = row.select('td')
    if len(cells) > 3:
        date=re.findall('\d+-\d+-\d+', cells[3].text.strip())
        time=re.findall('\d+:\d+:\d+.\d', cells[3].text.strip())
        values.append([
            date[0],
            time[0],
            cells[4].text.strip() + cells[5].text.strip(),
            cells[6].text.strip() + cells[7].text.strip(),
            cells[11].text.strip()
        ])
    
colnames=['date','time','latitude','longitude','region name']
df = pd.DataFrame(values,columns=colnames)
df.tail(20)
````

#### Count number of tweets by a given Twitter account.

You will need to include a ***try/except block*** for account names not found. 
<br>***Hint:*** the program should count the number of tweets for any provided account

````python
# This is the url you will scrape in this exercise 
# You will need to add the account credentials to this url
url = 'https://twitter.com/marca'
````

````python
#your code
try:
    html = requests.get(url).content
    soup = BeautifulSoup(html, "lxml")
    table=soup.find_all('span',{'class':'ProfileNav-value'})
    if table:
        info_tweets=[i.text.strip().split("\n") for i in table]
        tweets=info_tweets[0][0].replace("\xa0"," ")
    else:
        raise Exception('Account not found')
except Exception:
    raise
print(tweets)
````
